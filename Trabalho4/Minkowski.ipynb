```python
import pandas as pd
import numpy as np
```

```python
tabela = pd.read_csv('minkowski_results.csv')
```

```python
tabela
```

```python
def parse_points(point_string):
    if isinstance(point_string, str):
        points = []
        for point_str in point_string.split(';'):
            if point_str:
                x_str, y_str = point_str.strip('()').split(',')
                points.append((int(x_str), int(y_str)))
        return points
    return []

tabela['RobotPoints_list'] = tabela['RobotPoints'].apply(parse_points)
tabela['ObstaclePoints_list'] = tabela['ObstaclePoints'].apply(parse_points)
tabela['MinkowskiPoints_list'] = tabela['MinkowskiPoints'].apply(parse_points)

display(tabela.head())
```

```python
import math

def calculate_min_distance(polygon1, polygon2):
    min_distance = float('inf')
    for p1 in polygon1:
        for p2 in polygon2:
            distance = math.dist(p1, p2)
            min_distance = min(min_distance, distance)
    return min_distance

def calculate_distance_to_origin(polygon):
    origin = (0, 0)
    min_distance = float('inf')
    for p in polygon:
        distance = math.dist(p, origin)
        min_distance = min(min_distance, distance)
    return min_distance

tabela['MinDistance_Robot_Obstacle'] = tabela.apply(lambda row: calculate_min_distance(row['RobotPoints_list'], row['ObstaclePoints_list']), axis=1)
tabela['MinDistance_Minkowski_Origin'] = tabela['MinkowskiPoints_list'].apply(calculate_distance_to_origin)

display(tabela.head())
```

```python
distance_table = tabela[['RobotIndex', 'ObstacleIndex', 'MinDistance_Robot_Obstacle', 'MinDistance_Minkowski_Origin']].copy()
display(distance_table.head())
```
