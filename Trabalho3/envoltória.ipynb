# Bibliotecas

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# importando o dataset

```python
custo_computacional = pd.read_csv('custo_computacional.csv')
custo_computacional
```

```python
x_points = custo_computacional['num_pontos']
y_points = custo_computacional['tempo_ms']
```

# Plotando o gráfico

```python
plt.plot(x_points, y_points)
plt.xlabel("numero de pontos")
plt.ylabel("tempo em microssegundos")
plt.title("custo computacional em tempo")
plt.show()
```

```python
plt.plot(x_points, y_points)
plt.xlabel("numero de pontos")
plt.ylabel("tempo em microssegundos")
plt.title("custo computacional em tempo com pontos próximos")
plt.show()
```

```python
plt.plot(x_points, y_points)
plt.xlabel("numero de pontos")
plt.ylabel("tempo em microssegundos")
plt.title("custo computacional em tempo com pontos esparsos")
plt.show()
```
